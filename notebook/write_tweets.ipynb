```python
from string import punctuation as punct

import pandas as pd
```

```python
df = pd.read_pickle('../data/tweets/clean_tweets.pkl')
```

```python
file = open('../data/tweets/tweet_text.txt', 'w')

with open('../data/tweets/tweet_text.txt', 'w') as file:
    for tweet in df.clean_text:
        file.write(tweet.strip())
        if any(tweet.endswith(i) for i in punct):
            file.write(' ')
        else:
            file.write('. ')

        
print('done writing')
```
